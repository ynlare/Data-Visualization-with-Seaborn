```python
import pandas as pd 

import matplotlib.pyplot as plt 

import seaborn as sns 
```

```python
## Import dataset

history = pd.read_csv('/Users/Yedou/Desktop/Seaborn/museum_visitors.csv')
history
```

```python
history.head()
```

```python
# Data manipulation #

# Chose Date column as index

history.set_index('Date', inplace = True)

history.head()
```

```python
# Total of visitor by museum from 2014 to 2018

# The size of our chart

plt.figure(figsize=(15,6))
plt.xticks(rotation = 90)
# Representation of our line plot

sns.lineplot(data = history)

plt.title('Total of visitor by museum from 2014 to 2018')

plt.show()
```
